```python
#The structural model of schema TRP may be found in ../sql/create_trp_dimensions.sql


with engine.connect() as con1:
    q = "SELECT id, poi_tags_list FROM gtlab.sk_pois_activities_tripadvisor_0723_details order by poi_tags_list"
    df_ta_pois = pd.read_sql(q, con=con1)
    con1.close()

with engine.connect() as con1:
    q = "SELECT * from gtlab.trp_5_subgroups"
    df_tr_l5 = pd.read_sql(q, con=con1)
    con1.close()
    

with engine.connect() as con1:
    q = "SELECT * from gtlab.trp_4_groups"
    df_tr_l4 = pd.read_sql(q, con=con1)
    con1.close()
    
with engine.connect() as con1:
    q = "SELECT * from gtlab.trp_3_subclasses"
    df_tr_l3 = pd.read_sql(q, con=con1)
    con1.close()

with engine.connect() as con1:
    q = "SELECT * from gtlab.trp_2_classes"
    df_tr_l2 = pd.read_sql(q, con=con1)
    con1.close()

with engine.connect() as con1:
    q = "SELECT * from gtlab.trp_1_sources"
    df_tr_l1 = pd.read_sql(q, con=con1)
    con1.close()

#Clean the orginal lists of tags

for i, lt in zip(df_ta_pois.index,  df_ta_pois['poi_tags_list']):
    #print(lt)
    if lt == '{}':
        clean_plt = []
    else:
        new_lt = lt.replace('{', '').replace('}','').replace('"', '').split(',')
    
        clean_plt =  [x for x in new_lt if x != '']
    print(i, len(clean_plt), clean_plt)
    df_ta_pois.loc[i, 'clean_plt'] = str(clean_plt)
    df_ta_pois.loc[i, 'len_clean_plt'] = len(clean_plt)

#THIS OF COURSE CAN BE DONE VIA SIMPLE LIST COMPREHENSIONS, BUT I WANTED TO MAKE SURE REFERENCE TABLE ARE CORRECTLY DESIGNED...
#FASTER WAY INVOKE REFERENCE TABLES AS LIST AND LOOP THROUGH THEM
import ast
for i, id, clean_plt in zip (df_ta_pois.index, df_ta_pois.id, df_ta_pois.clean_plt):
    l_id_tr_l5 = []
    l_id_tr_l4 = []
    for x in ast.literal_eval(clean_plt):
        for id_subgroup, name_subgroup, id_group  in list(zip(df_tr_l5['id'], df_tr_l5['name'], df_tr_l5['group_id'] )):
            if x.lstrip() == name_subgroup:
                l_id_tr_l5.append(id_subgroup)
                l_id_tr_l4.append(id_group)
    df_ta_pois.loc[i, 'l_id_tr_l5'] = str(l_id_tr_l5)
    df_ta_pois.loc[i, 'l_id_tr_l4'] = str(l_id_tr_l4)


for i, id, l_id_tr_l4 in zip (df_ta_pois.index, df_ta_pois.id, df_ta_pois.l_id_tr_l4):
    l_id_tr_l3 = []
    for x in ast.literal_eval(l_id_tr_l4):
        for id_group, id_subclass in list(zip(df_tr_l4['id'], df_tr_l4['subclass_id'])):
            if id_group != None:
                if int(x) == int(id_group):
                    l_id_tr_l3.append(id_subclass)
    df_ta_pois.loc[i, 'l_id_tr_l3'] = str(l_id_tr_l3)


for i, id, l_id_tr_l3 in zip (df_ta_pois.index, df_ta_pois.id, df_ta_pois.l_id_tr_l3):
    l_id_tr_l2 = []
    for x in ast.literal_eval(l_id_tr_l3):
        for id_subclass, id_class in list(zip(df_tr_l3['id'], df_tr_l3['class_id'])):
            if id_subclass != None:
                if int(x) == int(id_subclass):
                    l_id_tr_l2.append(id_class)
    df_ta_pois.loc[i, 'l_id_tr_l2'] = str(l_id_tr_l2)
    

for i, id, l_id_tr_l2 in zip (df_ta_pois.index, df_ta_pois.id, df_ta_pois.l_id_tr_l2):
    l_id_tr_l1 = []
    for x in ast.literal_eval(l_id_tr_l2):
        for id_class, id_source in list(zip(df_tr_l2['id'], df_tr_l2['sources_id'])):
            if id_class != None:
                if int(x) == int(id_class):
                    l_id_tr_l1.append(id_source)
    df_ta_pois.loc[i, 'l_id_tr_l1'] = str(l_id_tr_l1)
```
